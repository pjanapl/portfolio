Poniewa≈º mamy dane o ratingu w DF reviews dotyczƒÖce powy≈ºszych wpis√≥w, uzupe≈Çniƒô te wpisy o dane z DF reviews.

```python
missing_rating_url_list = list(
    products[
        (products['url'].isin(reviews['product_url'].unique())) &
        (products['rating'].isna())
    ]['url']
)

rating_from_reviews = reviews.groupby('product_url')['rating'].mean().reset_index()
rating_from_reviews[rating_from_reviews['product_url'].isin(missing_rating_url_list)]

```

Rzeczywi≈õcie dane o ratingu dla tych 26 wpis√≥w istniejƒÖ. Wystarczy wiƒôc przepisaƒá warto≈õci z pomocniczego DF (rating_from_reviews) do DF products.

```python
merged = products.merge(
    rating_from_reviews[['product_url', 'rating']],
    how = 'left',
    left_on = 'url',
    right_on = 'product_url',
    suffixes = ('', '_from_reviews')
)

merged['rating'] = merged['rating'].combine_first(merged['rating_from_reviews'])
merged = merged.drop(columns=['rating_from_reviews', 'product_url'])

merged[merged['url'].isin(missing_rating_url_list)][['rating', 'url']]
```

```python
products = merged.reset_index(drop = True)
products.info()
```

26 brakujƒÖcych rating√≥w zosta≈Ço uzupe≈Çnione. Usunƒô kolumnƒô reviews, poniewa≈º szczeg√≥≈Çowe dane przechowujƒô w DF reviews.


```python
products = products.drop(columns = ['reviews'])
products.info()
```

Mamy wiƒôc 856 rating√≥w unikalnych produkt√≥w w DF reviews i 891 w DF products. W DF reviews brakuje wiƒôc danych 35 a nie 9 produkt√≥w (r√≥≈ºnica wynika z tych 26 produkt√≥w, kt√≥re nie mia≈Çy ratingu w DF products).

```python
reviews_missing = pd.DataFrame(pd.read_pickle('../data/reviews_missing_2.pkl'))
reviews_missing.head()
```

```python
reviews_combined = pd.concat([reviews, reviews_missing], ignore_index=True)
reviews_combined['product_url'].nunique()
```

Dane w obu DF sƒÖ teraz zgodne - istniejƒÖ dane o ocenie u≈ºytkownik√≥w dla 891 produkt√≥w. Do dalszej analizy we≈∫miemy tylko produkty zawierajƒÖce oceny u≈ºytkownik√≥w.

```python
reviews_combined.info()
```

Stworzƒô DF zawierajƒÖcy dane z DF reviews po≈ÇƒÖczone z niekt√≥rymi danymi z DF products - id, markƒÖ, nazwƒÖ produktu, docelowƒÖ kategoriƒÖ, aktualnƒÖ cenƒÖ i warto≈õciƒÖ przeceny. Usunƒô natomiast product_url. Wynikowo bƒôdƒô mia≈Ç d≈Çugi DF, w kt√≥rym dla ka≈ºdego komentarza bƒôdƒô mia≈Ç kluczowe informacje o produkcie.

```python
reviews_extended = reviews_combined.merge(
    products[['product_id', 'name', 'url', 'brand', 'current_price_pln', 'discount_pct', 'top_level_category']],
    how = 'left',
    left_on = 'product_url',
    right_on = 'url'
)
reviews_extended.drop(columns = ['product_url', 'url'], inplace = True)
# Zmiana formatu daty
reviews_extended['date'] = pd.to_datetime(reviews_extended['date'], format='%d.%m.%Y')

reviews_extended.info()
```
