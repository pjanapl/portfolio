Dane o produktach (DataFrame products) posiadają 865 ocen produktów a nowe dane (DataFrame reviews) 856. Przeba więc przyjrzeć się tej różnicy.

```python
# Liczba produktów, których url nie znajduje się na liście reviews, i które nie są puste
len(products[
    (~products['url'].isin(reviews['product_url'].unique())) &
    (~products['rating'].isna())
])
```

Takich produktów jest 35, więc na liście muszą znajdować się też produkty, które znajdują się w DataFrame reviews a w DataFrame products mają pusty rating.

```python
len(products[
    (products['url'].isin(reviews['product_url'].unique())) &
    (products['rating'].isna())
])
```

```python
products[
    (products['url'].isin(reviews['product_url'].unique())) &
    (products['rating'].isna())
][['rating', 'url']]
```

Ponieważ mamy dane o ratingu w DF reviews dotyczące powyższych wpisów, uzupełnię te wpisy o dane z DF reviews.

```python
missing_rating_url_list = list(
    products[
        (products['url'].isin(reviews['product_url'].unique())) &
        (products['rating'].isna())
    ]['url']
)

rating_from_reviews = reviews.groupby('product_url')['rating'].mean().reset_index()
rating_from_reviews[rating_from_reviews['product_url'].isin(missing_rating_url_list)]

```

Rzeczywiście dane o ratingu dla tych 26 wpisów istnieją. Wystarczy więc przepisać wartości z pomocniczego DF (rating_from_reviews) do DF products.

```python
merged = products.merge(
    rating_from_reviews[['product_url', 'rating']],
    how = 'left',
    left_on = 'url',
    right_on = 'product_url',
    suffixes = ('', '_from_reviews')
)

merged['rating'] = merged['rating'].combine_first(merged['rating_from_reviews'])
merged = merged.drop(columns=['rating_from_reviews', 'product_url'])

merged[merged['url'].isin(missing_rating_url_list)][['rating', 'url']]
```

```python
products = merged.reset_index(drop = True)
products.info()
```

26 brakujących ratingów zostało uzupełnione. Usunę kolumnę reviews, ponieważ szczegółowe dane przechowuję w DF reviews.


```python
products = products.drop(columns = ['reviews'])
products.info()
```

Mamy więc 856 ratingów unikalnych produktów w DF reviews i 891 w DF products. W DF reviews brakuje więc danych 35 a nie 9 produktów (różnica wynika z tych 26 produktów, które nie miały ratingu w DF products).

```python
# Liczba produktów, których url nie znajduje się na liście reviews, i które nie są puste w DF products
len(products[
    (~products['url'].isin(reviews['product_url'].unique())) &
    (~products['rating'].isna())
])
```

```python
reviews_missing = pd.DataFrame(pd.read_pickle('../data/reviews_missing_2.pkl'))
reviews_missing.head()
```

```python
reviews_combined = pd.concat([reviews, reviews_missing], ignore_index=True)
reviews_combined['product_url'].nunique()
```

Dane w obu DF są teraz zgodne - istnieją dane o ocenie użytkowników dla 891 produktów. Do dalszej analizy weźmiemy tylko produkty zawierające oceny użytkowników.

```python
reviews_combined.info()
```

Stworzę DF zawierający dane z DF reviews połączone z niektórymi danymi z DF products - id, marką, nazwą produktu, docelową kategorią, aktualną ceną i wartością przeceny. Usunę natomiast product_url. Wynikowo będę miał długi DF, w którym dla każdego komentarza będę miał kluczowe informacje o produkcie.

```python
reviews_extended = reviews_combined.merge(
    products[['product_id', 'name', 'url', 'brand', 'current_price_pln', 'discount_pct', 'top_level_category']],
    how = 'left',
    left_on = 'product_url',
    right_on = 'url'
)
reviews_extended.drop(columns = ['product_url', 'url'], inplace = True)
# Zmiana formatu daty
reviews_extended['date'] = pd.to_datetime(reviews_extended['date'], format='%d.%m.%Y')

reviews_extended.info()
```
